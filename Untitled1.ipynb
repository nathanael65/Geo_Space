```python
import pandas as pd
import geopandas as gpd
```

```python
# Ces differents codes ci-dessous permettent d'afficher les données du fichier excell point_eau.xlsx(code commenté), ainsi que le fichier CSV point_eaux.csv

#path = r"/content/Point_eau.xlsx"
#pd.read_excel(path)
path = r"/content/Point_eau.csv"
csv = pd.read_csv(path)
csv
```

```python
# Ce code permet d'ajouter une colonne géométrique au données affichés précédemment
df_csv=gpd.GeoDataFrame(csv,geometry=gpd.points_from_xy(csv.longitude	, csv.lLatitude))
df_csv
```

```python
# ce code est important pour afficher les différents points
df_csv.plot("Nom")
```

```python
# l'intruction suivante permet d'afficher les ponts avec une légende
df_csv.plot(column="Nom",legend=True)
```

```python
# Ce code permet d'afficher la carte des provinces du Burkina
carte_provinces = gpd.read_file(r"/content/ADM_Province.shp")
carte_provinces.plot(cmap="hsv", edgecolor="black", figsize=(10,10))
```

```python
# Ce code permet d'afficher la carte des provinces du Burkina avec les point du fichier Point_d'eau.csv
axis=carte_provinces.plot(edgecolor="black", categorical=True, k=10,cmap="OrRd",
legend= True , figsize=(15,15))
df_csv.plot(ax=axis, color="blue")
```

```python
# ce code permet d'afficher la carte des points d'eaux avec un bufffet de 5 km (5km autour de chaque point)
df_csv.crs="epsg:4326"
from matplotlib import *
df_csv.to_crs(3857)
df_csv['buffered']=df_csv['geometry'].buffer(0.5)
ax=df_csv['buffered'].plot(alpha=0.3)
df_csv.plot(ax=ax, color='red')
```
